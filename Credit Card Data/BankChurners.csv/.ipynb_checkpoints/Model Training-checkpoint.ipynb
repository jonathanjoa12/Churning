### Neural Networks

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
network = Sequential()
network.add(Dense(32,
                 input_shape = (Xsm_train.shape[1],),
                 activation = 'relu'))
network.add(Dense(1, activation = 'sigmoid'))
```

```python
network.compile(loss='binary_crossentropy', optimizer='adam', metrics = ['f1_score'])
```

```python
history = network.fit(Xsm_train, ysm_train, validation_data = (Xs_test, y_test), epochs = 10, batch_size = 512, verbose = 0)
```

```python
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.plot(history.history['loss'], label = 'loss')
plt.legend();
```
